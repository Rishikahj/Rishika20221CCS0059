```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,5]
plt.plot(x,y)
plt.title("info")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
x=[10,20,30,40]
mylabels=["Apple","Banana","cherries","dragonfruit"]
mycolors=["red","Yellow","blue","pink"]
d=[0.2,0,0.5,0.2]
plt.pie(x,labels=mylabels,colors=mycolors,explode=d,shadow=True)
plt.show()
```

```python
x=[1,3,5,7]
y=[6,7,8,9]
colors=["g","r","b","orange"]
sizes=[100,200,300,400]
plt.scatter(x,y,color=colors,s=sizes,alpha=0.8,marker="*",edgecolor="g")
plt.title("scatter",fontsize=30)
plt.xlabel("x axis",fontsize=10)
plt.ylabel("y axis",fontsize=10)
plt.title("Bar plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3]
y=[4,5,5]
plt.subplot(2,3,1)
plt.plot(x,y,color="r")
plt.title("info")
plt.xlabel("x axis")
plt.ylabel("y axis")

x=[10,20,30,40]
plt.subplot(2,3,2)
mylabels=["Apple","Banana","cherries","dragonfruit"]
mycolors=["red","Yellow","blue","pink"]
d=[0.2,0,0.5,0.2]
plt.pie(x,labels=mylabels,colors=mycolors,explode=d,shadow=True)
x=[1,3,5,7]
y=[6,7,8,9]
plt.subplot(2,3,3)
plt.title("bar plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.bar(x,y)

x=[1,3,5,7]
y=[6,7,8,9]
plt.subplot(2,3,4)
colors=["g","r","b","orange"]
plt.bar(x,y,color=colors)
plt.xlabel("x axis",fontsize=10)
plt.ylabel("y axis",fontsize=10)
plt.title("Bar plot")

x=[1,3,5,7]
y=[2,5,6,8]
plt.subplot(2,3,5)
colors=["g","r","b","orange"]
sizes=[100,200,300,400]
plt.scatter(x,y,color=colors,s=sizes,alpha=0.8,marker="*",edgecolor="g")
plt.xlabel("x axis",fontsize=10)
plt.ylabel("y axis",fontsize=10)
plt.title("scatter plot",fontsize=15)

```
